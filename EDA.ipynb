```python
# Remover linhas em branco e ou vazias
dataset.dropna(how="all", inplace=True)
```

```python
# Validando algumas hipóteses
# Uma hipótese se define como uma suposição ou uma possibilidade considerada válida.
# Vamos avaliar as seguintes hipóteses:

# H1 – Existe alguma predominância de sexo nas reclamações? Ou melhor, qual quantidade de reclamações por sexo?
# H2 – Qual a faixa etária média dos consumidores?
# H3 – Qual o grupo de problema mais comum e o segmento de mercado mais afetado?
# H4 – Existe alguma correlação entre o tempo de resposta e a nota do consumidor?
# H5 – Qual a proporção de reclamações registradas e que não são resolvidas ?
```

```python
# Qual o número de reclamações por Estado?
total_por_UF = dataset.groupby('UF')['UF'].count()
print(total_por_UF)
```

```python
# H1 – Existe alguma predominância de sexo nas reclamações? Ou melhor, qual quantidade de reclamações por sexo?
qtd_por_sexo = dataset.groupby('Sexo')['Sexo'].count()
print(qtd_por_sexo)
```

```python
# outro método para efetuar a contagem por agrupamento do sexo
dataset[u'Sexo'].value_counts()
```

```python
# H2 – Qual a faixa etária média dos consumidores?
faixa_etaria = dataset.groupby('Sexo')['Faixa Etária'].value_counts()
print(faixa_etaria)
```

```python
# Para responder a hipótese H3 vamos concentrar nos atributos e Grupo Problema e Segmento de Mercado.
dataset['Grupo Problema'].value_counts()
```

```python
# Qual o segmento de mercado mais afetado ?
# Queremos saber qual o segmento de mercado com o maior número de reclamações do grupo de problema Cobrança e Contestação.

segmento = dataset[dataset[u'Grupo Problema']==u'Cobrança / Contestação']
segmento['Segmento de Mercado'].value_counts()
```

```python
fig, ax = plt.subplots()
y_pos = np.arange(len(segmento['Segmento de Mercado'].value_counts()))
values = segmento['Segmento de Mercado'].value_counts()
segmentos = segmento['Segmento de Mercado'].unique()
ax.barh(y_pos, values, align='center', color='blue')
ax.set_yticks(y_pos)
ax.set_yticklabels(segmentos)
ax.invert_yaxis()
ax.set_xlabel('Reclamacoes')
ax.set_title('Cobranca e Contestacao por Segmento de Mercado')
plt.show()
```

```python
# Observe que a grande maioria de empresas com reclamações registradas com problemas do tipo:
# Bancos, Financeiras e Administradoras de Cartão e Empresas de intermediação de serviços/Negócios.
# Vamos destacar os nomes das empresas delinquentes:
fig, ax = plt.subplots()
y_pos = np.arange(len(segmento['Nome Fantasia'].value_counts()[:20]))
values = segmento['Nome Fantasia'].value_counts()[:20]
segmentos = segmento['Nome Fantasia'].unique()[:20]
ax.barh(y_pos, values, align='center', color='green')
ax.set_yticks(y_pos)
ax.set_yticklabels(segmentos)
ax.invert_yaxis()
ax.set_xlabel('Reclamacoes')
ax.set_title('Rank de Empresas')
plt.show()
```

```python
# Existe alguma correlação entre o tempo de resposta e a nota do consumidor ?
# A hipótese H4 surge da idéia de que o tempo que a empresa leva para responder o consumidor pode impactar na sua nota final.
# Será que tem fundamento ?
# Para tentar validar essa suposição utilizaremos uma técnica chamada Coeficiente de Correlação de Pearson.
# A técnica basicamente mede o grau de correlação e sua direção entre duas variáveis de escala métrica.
# O coeficiente assume valores entre -1  e 1 onde o valor 1 indica uma correlação perfeitamente positiva entre as duas variáveis,
# -1 indica uma correlação perfeitamente negativa e o valor 0 representa que não existe correlação linear entre as variáveis.
# Calculando a correlação e visualizando os dados
```

```python
# Qual a proporção de reclamações registradas e que não são resolvidas ?
'''
Já vimos que o número de reclamações registradas no sistema é grande mas fica a pergunta; 
Quantas dessas reclamações registradas não são efetivamente resolvidas?
Essa é a provocação da hipótese H5 – Qual a proporção de reclamações registradas e que não são resolvidas ?
'''
############################################
# Vamos selecionar o total de reclamações e 
# a quantidade de reclamações não resolvidas por segmento de mercado das empresas através do código abaixo.

x = dataset.groupby(u'Segmento de Mercado')[u'Segmento de Mercado'].count()
#Filtrando os segmentos de mercado com mais de 20 reclamações
x = x[x > 20]
df = dataset[dataset[u'Segmento de Mercado'].isin(x.keys())]
df = df[df[u'Avaliação Reclamação']==u'Não Resolvida']
x1 = df.groupby(u'Segmento de Mercado')[u'Segmento de Mercado'].count()
```

```python
# Com os dados selecionados podemos renderizar o gráfico com o código abaixo.
sns.set(style="whitegrid")
f, ax = plt.subplots(figsize=(8, 10))
sns.set_color_codes("pastel")
sns.barplot(x=x.values, y=x.keys(), label="Total de reclamacoes", color="b")
sns.set_color_codes("muted")
sns.barplot(x=x1.values, y=x1.keys(),label="Reclamacoes nao resolvidas", color="r")
ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(ylabel="", xlabel="Reclamacoes e suas situacoes")
sns.despine(left=False, bottom=True)
```
